## Challenge - Square List
Generate a list of squares of all numbers between 1 to 100. Do it in one line of code.
Output: [1, 4, 9, 16, 25, 36, ..., 10000]

```python
squareList=[i**2 for i in range(1,101)]
print(squareList)
```

## Even List

```python
resultOfEven=[i**2 for i in range(1,101) if i%2==0]
print(resultOfEven)
```

## Using map function

```python
def square(x):
    return x**2
result=list(map(square,range(1,101)))
print(result)
    
```

## Challenge - Even List
Given a list of random numbers, filter out all even numbers from it. Do it in one line of code.
Input: [27, 24, 21, 46, 48, 6, 25, 38, 10, 40]
Output: [24, 46, 48, 6, 38, 10, 40]

```python
Input= [27, 24, 21, 46, 48, 6, 25, 38, 10, 40]
output=[i for i in Input if i%2==0]
print(output)
```

# Tuples
Tuples are used to store multiple items in a single variable.
A tuple is a collection which is ordered and immutable (unchangeable).
Tuples are written with round brackets.
Tuple items are indexed, the first item has index [0], the second item has index [1] etc.
Tuples can be created with tuple constructor - tuple()
Tuples can also be created by simply writing multiple values with comma separation.
Tuples are processed faster than lists in Python.

```python
continents = ("Africa", "Antarctica", "Asia", "Europe", "North America", "South America", "Australia")
# read only list
```

```python
print(type(continents))
```

```python
def checkSort(x):# time complexity=O(n)
    for i in range(len(x)-1):
        if x[i]<=x[i+1]:
            continue
        else:
            return -1

    return 1
            
check=(1,4,10,12,114,18,20)
rem=checkSort(check)
if rem==1:
    print('sorted')
else:
    print("Not sorted")
    
```

## Challenge - Write a function which returns the product of all its arguments.
The user should be able to call the function like
product(2, 3, 9, 1)  # 54
product(4, 2)        # 8

```python
def product(*args):
    p=1
    for i in args:
        p=p*i
    return p
print(product(1,2,3,4,5))
```
