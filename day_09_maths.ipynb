```python
def combinations(n,r):
    num=math.factorial(n)
    den=math.factorial(r)*math.factorial(n-r)
    return num/den
print(combinations(10,4))
```

## Decimal to Binary

```python
def decimalTObinary(num):
    a=[]
    temp=num
    while temp>=1:
        a.append(str(temp%2))
        temp=temp//2
    a=a[::-1]
    final=int("".join(a))
    print(final)

decimalTObinary(25)
    
```

## Binary to Decimal

```python
def BinToDec(bin):
    '''
    binary to decimal conversion
    '''
    sum=0
    temp_bin=bin
    i=0
    while temp_bin >=1:
        sum=sum+(temp_bin%10)*pow(2,i)
        i=i+1
        temp_bin=temp_bin//10
    return sum
n=int(input("Enter the number in binary-"))
print(BinToDec(n))
```

### Number of Trailing zeros in the factorial of a number
Given an integer A, return the number of trailing zeroes in A!, i.e., factorial of A.\
Note: Your solution should run in logarithmic time complexity

```python
def findTrailingZeros(n):
    factN=math.factorial(n)
    temp=factN
    count =0
    while temp>=1:
        if temp%10==0:
            count+=1
            temp=temp//10
        else:
            break
    return count 
findTrailingZeros(1000)
```

## optimised approach

```python
def findZeros(n):
    factN=math.factorial(n)
    fiveCounter=0
    temp=factN
    while temp%5==0:
        fiveCounter+=1
        temp=temp//5
    return fiveCounter
findZeros(1000)
```

## optimising factorial calculation

```python
def countTrailingZ(n):
    countFives =0
    temp=n
    while temp>=1:
        countFives+=temp//5
        temp=temp//5
    countTwo=0
    temp=n
    while temp>=1:
        countTwo+=temp//2
        temp=temp//2
    if countTwo<=countFives:
        return countTwo
    else:
        return countFives
countTrailingZ(1000)
        
    
```
