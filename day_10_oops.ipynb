# OOPS

Object-oriented programming aims to implement real-world entities like inheritance, hiding, polymorphism, etc in programming. \
The main aim of OOP is to bind together the data and the functions that operate on them so that no other part of \
the code can access this data except that function.   \
Classes and objects are two main aspects of OOPs.\
Classes are the user-defined blueprints that help us create an object.\
Objects are the instances of a particular class.   \



```python
class Car:
    name="BMW S series"
    color="Woody Brown"
    age=1
    def describe(self):
        print("It works")
```

```python
BMW=Car() # constructs and returns an object
```

```python
print(BMW.name)
```

```python
suzuki=Car()
suzuki.name="Suzuki 16"
suzuki.age=5
suzuki.color="Purple"
suzuki.fuel="EV" # we can add new properties as we want
suzuki.describe()
```

```python
print(BMW.name,BMW.color,BMW.age)
```

```python
print(suzuki.name,suzuki.color,suzuki.age,suzuki.fuel)
```

```python
class bikes:
    def describe(self):
        print("Bike name-",self.name)
        print("Bike color-",self.color)
        print("Bike price-",self.price)
Royal_Enfield=bikes()
Royal_Enfield.name=" Bullet 350 ES"
Royal_Enfield.color="Bloody Red"
Royal_Enfield.price=196635
Harley_Davidson=bikes()
Harley_Davidson.name="Fat Bob"
Harley_Davidson.color="Bloody Red"
Harley_Davidson.price=3556985
```

```python
Royal_Enfield.describe()
```

```python
Harley_Davidson.describe()
```

```python
class car:
    def __init__(self,name,color,age):
        self.name=name
        self.color=color
        self.age=age
    def isBanned(self):
        if self.age>10:
            print("Banned")
        else:
            print("Not Banned")
bmw=car("BMW S series","Woody Brown",5)
bmw.isBanned()
```
