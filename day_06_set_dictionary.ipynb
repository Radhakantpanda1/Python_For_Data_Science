# SETS
Set is an unordered collection of items, where every element is unique and immutable. However, the set itself is mutable.
There are two ways of creating a set:
Putting all the elements inside curly braces “{ }”
Using set() method.
Adding elements to set - add(), update()
Removing elements from set - remove(), discard()
Python sets allow fast insertion, deletion and searching.

```python
A={1,6,8,45,7,9,6,22}
print(A)
```

### Insertion in list O(n)
### Deletion in list O(n)
### Searching In list O(n)

### Insertion deletion and searching takes O(1) time in sets

```python
A.add(101)
# if value is not present it gives error
```

```python
A.update(["hELLO",True])
print(A)
```

```python
A.remove(101)
print(A)
```

```python
A.discard(101)
# if value is not present it doesn't gives error
```

Given a string A consisting of lowercase characters.

Check if characters of the given string can be rearranged to form a palindrome.

Return 1 if it is possible to rearrange the characters of the string A such that it becomes a palindrome else return 0.

### solution

```python
"""
def solve(self, A):
        frequency={}
        for character in A:
            if character in frequency:
                frequency[character]+=1
            else:
                frequency[character]=1
        odd_frequency_counter=0
        for freq in frequency.values():
            if freq % 2 ==1:
                odd_frequency_counter+=1
        if odd_frequency_counter<=1:
            return 1
        else: 
            return 0
"""
```

```python
def solve( A):
        frequency={}
        for character in A:
            if character in frequency:
                frequency[character]+=1
            else:
                frequency[character]=1
        odd_frequency_counter=0
        for freq in frequency.values():
            if freq % 2 ==1:
                odd_frequency_counter+=1
        if odd_frequency_counter<=1:
            return 1
        else: 
            return 0
```
