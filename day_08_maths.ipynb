## Logarithm
The exponent or power to which a base must be raised to yield a given number.\
Expressed mathematically, x is the logarithm of n to the base 1  b if bx = n, in which case one writes x = logb n.\
For example, 23 = 8; therefore, 3 is the logarithm of 8 to base 2, or 3 = log2 2  8

```python
firstNum=int(input("Enter the number"))
secondNum=int(input("Enter the number"))
if firstNum>secondNum:
    firstNum,secondNum=secondNum,firstNum
while firstNum!=0:
    temp=firstNum
    firstNum=secondNum%firstNum
    secondNum=temp
print("GCD=",secondNum)
```

```python
def getGCD(firstNum,secondNum):
    while firstNum!=0:
        firstNum,secondNum=secondNum%firstNum,firstNum
    return secondNum
```

```python
getGCD(2,3)
```

# LCM

```python
def getLcm(firstNum,secondNum):
    gcd=getGCD(firstNum,secondNum)
    lcm=(firstNum*secondNum)/gcd
    return lcm
```

```python
getLcm(2,3)
```

```python
math.lcm(5,6)
```

# Prime Number using Square Root Method

```python
import math
def checkPrime(num):
    '''
    checks prime
    '''
    if num==1:
        return 0
    isPrime=True
    for i in range(2,int(math.sqrt(num))+1):
        if num%i==0:
            isPrime=False
            return 0
    return 1
num=int(input("Enter the number-"))
status=checkPrime(num)
if status==1:
    print("Prime")
else:
    print("Not a Prime")      
```
