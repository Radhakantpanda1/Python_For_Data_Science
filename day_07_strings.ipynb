## 1ans

```python
def findMatrixSum(matrixA,matrixB):
    finalMatrix=[]
    for row in range(len(matrixA)):
        temp=[]
        for column in range(len(matrixA[row])):
            temp.append(matrixA[row][column]+matrixB[row][column])
        finalMatrix.append(temp)
    return finalMatrix
matrixA=[[1,2],[3,4,5]]
matrixB=[[5,6],[7,8,5]]
final=findMatrixSum(matrixA,matrixB)
print(final)
```

## 2ans

```python
def findMatrixDiff(matrixA,matrixB):
    finalMatrix=[]
    for row in range(len(matrixA)):
        temp=[]
        for column in range(len(matrixA[row])):
            temp.append(matrixA[row][column]-matrixB[row][column])
        finalMatrix.append(temp)
    return finalMatrix
matrixA=[[1,2],[3,4,5]]
matrixB=[[5,6],[7,8,5]]
final=findMatrixDiff(matrixA,matrixB)
print(final)
```

# Function kwargs(keyword arguments)

```python
def userDetails(name,age,**kwargs):
    print(name,age)
    print(kwargs)# kwargs is a dictionary type
```

```python
userDetails('Radhakant Panda',22,city="Aska",state="Odisha")
```

```python
def hogwarts(housel, house2, house3, house4, *args, **kwargs):
    print("Houses are ", housel, house2, house3, house4)

    print()
    print("Args :", args)

    print()
    print("Kwargs :", kwargs)
```

```python
hogwarts(
    "Gryffindor",
    "Ravenclaw",
    "Hufflepuff",
    "Slytherin",
    "Severus Snape",
    "Horace Slughorn",
    "Remus Lupin",
    headmaster="Albus Dumbledore",
    deputy_headmistress="Minerva McGonagall"
)
```
